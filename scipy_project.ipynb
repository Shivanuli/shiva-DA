```python
np.random.seed(0)
df4 = pd.DataFrame({
    'Year': np.random.choice(['2020', '2021', '2022'],size=18),
    'Amount':np.random.normal(loc=10, scale=2, size=18)#loc=>mean,scale=>standard deviation
})
print(df4)
```

```python
#claculate mean amount for each year
mean_score_per_group = df4.groupby('Year')['Amount'].mean()
print(mean_score_per_group)
```

```python
#generate random year selections
choices = ['2020', '2021', '2022']
size = 18
year_column = np.random.choice(choices, size=size)
print(year_column)
```

```python
#calculate amount values
mean = 10
std = 2
size = 18
Amount_values = mean + std * np.random.randn(size)
print(Amount_values)
```

```python
# hypothesis testing (t-test)
group_A = df4[df4['Year'] == '2020']['Amount']
group_B = df4[df4['Year'] == '2021']['Amount']
t_stat, p_value =stats.ttest_ind(group_A, group_B)
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
#prediction
if p_value < 0.05:
  print("reject the null hypothesis")
else:
  print("fail to reject the null hypothesis")
```

```python
#ANOVA (one-way ANOVA)
f_stat, p_value = stats.f_oneway(df4[df4['Year']== '2020']['Amount'],
                                 df4[df4['Year']== '2021']['Amount'],
                                 df4[df4['Year']== '2022']['Amount'])
print("F-statistic:",f_stat)
print("p-value:",p_value)
```
